### Questão 2
 Considere o sinal 𝑥(𝑡) de tempo contínuo dado conforme a Figura 1 abaixo.



---

(a) Determine, de forma analítica, uma expresssão matemática para o sinal 𝑥(𝑡) para todo o t 

(b) Plote o sinal 𝑥(𝑡).

```python

import numpy as np
import matplotlib.pyplot as plt

# define a função degrau unitário

def u(t):
    return np.where(t >= 0, 1.0, 0.0)

# domínio de tempo para visualização

t = np.linspace(-5, 5, 3000)

# sinal x(t)
x = 2 * (u(t + 1.5) - u(t)) + 2 * np.exp(-t/2) * (u(t) - u(t-3))


def plota_sinal(sinal, mensagem):
    
   
    plt.figure(figsize=(9, 4))

    plt.plot(t, sinal, color='red',  linewidth=2, label="x(t)")

    # força mostrar o eixo zero completo
    plt.axhline(0, color="black", linewidth=1)
    plt.axvline(0, color="gray", linewidth=1)


    plt.yticks([0, 2])

    plt.title(f"{mensagem}")
    plt.xlabel("t")
    plt.ylabel("x(t)")

    plt.grid(True, linestyle="--", alpha=0.5)
    plt.legend()
    plt.ylim(-0.2, 2.4)
    plt.xlim(-6, 7)

    plt.show()
    
plota_sinal(x, "Sinal x(t)")
```

(c) Plote os seguintes sinais:

    (i) 𝑥(−𝑡)


```python
def sinal(t):
    return (2 * np.logical_and(t >= -1.5, t <= 0)) +  (2 * np.exp(-t/2) * np.logical_and(t > 0, t <= 3))


t = np.linspace(-5, 5, 3000)
b = sinal(-t)

    
plota_sinal(b, "sinal invertido no tempo")
```

    (ii) 𝑥(𝑡 + 4)

```python
t = np.linspace(-6, 6, 3000)
c = sinal(t+4)
plota_sinal(c, "Sinal deslocado no tempo: x(t+4)")
```

    (iii) 𝑥(𝑡 − 2)


```python
d = sinal(t-2)
plota_sinal(d, "Sinal deslocado no tempo: x(t-2)")
```

    (iv) 𝑥(2𝑡)


```python
e = sinal(2*t)
plota_sinal(e, "Sinal comprimido no tempo")
```

    (v) 𝑥(𝑡/2)

```python
t = np.linspace(-6, 8, 3000)
f = sinal(t/2)
plota_sinal(f, "Sinal ampliado no tempo: x(t/2)")
```
