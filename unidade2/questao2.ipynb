# Questão 2

Utilizando o Python, obtenha o gráfico da resposta em magnitude (em dB) para esse filtro
considerando que R = 10 kΩ e C = 10 nF. Verifique que o circuito corresponde à um filtro passa
baixas.

--- 

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

R = 10e3      # 10 kΩ
C = 10e-9     # 10 nF

num = [1]
den = [R*C, 1]
Hs = signal.TransferFunction(num, den)

f = np.logspace(1, 5, 1000)
w = 2*np.pi*f

_, Hw = signal.freqresp(Hs, w)

H_dB = 20*np.log10(np.abs(Hw))



plt.figure(figsize=(8,5))
plt.semilogx(f, H_dB, 'k')
plt.xlabel('$f$ (Hz)')
plt.ylabel('$|H(f)|$ (dB)')
plt.title('Resposta em Magnitude do Filtro Passa-Baixas RC')
plt.legend()
plt.show()
```

Nota-se que trata-se de um filtro passa baixas, pois para baixas frequências o ganho fica próximo de 0db, enquanto que para frequências mais altas, o ganho cai rapidamente.
