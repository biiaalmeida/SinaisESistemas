# Questão 4

Um determindado sistema linear e invariante no tempo possui uma resposta ao impulso $h(t)$ dada por $h(t) = e^{-2t}u(t)$.

---

a) Plote a resposta ao impulso do sistema para $0 \leq t \leq 3$.

```python
import numpy as np
import matplotlib.pyplot as plt

t_inicio = 0
t_fim = 3
npoints = 1000

t = np.linspace(t_inicio, t_fim, npoints)

h = np.exp(-2*t)
plt.plot(t, h)
plt.title('Resposta ao impulso do sistema')
plt.xlabel('t')
plt.ylabel('h(t)')
plt.grid()
plt.show()
```

---

b) O sistema é estável? O sistema é causal? Justifique suas respostas.

**Resposta:** Como a resposta ao impulso é absolutamente integrável, o sistema é estável. Além disso, O sistema é causal, pois para valores menores que zero, a saída é 0.

---

c) Considere que o seguinte sinal:

$$x(t) = u(t) - u(t-2)$$

é aplicado na entrada do sistema. Determine, de forma analítica, a expressão da saída  $y(t)$ para todo t utilizando o método gráfico da convolução.

```python
t = np.linspace(0, 6, 1000)

y = np.where(
    (t >= 0) & (t <= 2),
    0.5 - 0.5*np.exp(-2*t),
    np.where(
        t > 2,
        0.5*np.exp(4)*np.exp(-2*t) - 0.5*np.exp(-2*t),
        0
    ))

plt.figure(figsize=(12, 3))
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Resultado da convolução gráfica')
plt.grid(True)

plt.show()
```

---

d) resolva o item (c) de forma númerica. Compare o resultado numérico com o resultado analítico, obtidido no item (c). Uma forma de comparar consiste em plotar, em um mesmo gráfico, o sinal da saída obtida da forma analítica e o sinal de saída de forma numérica.

```python
x = np.where((t >= 0) & (t < 2), 1, 0)

y = np.convolve(x, h) * (t[1] - t[0])
t_y = np.linspace(0, 2*t_fim, len(y))
t_x = np.linspace(t_inicio, t_fim, npoints)

fig, axs = plt.subplots(3, 1, figsize=(10, 8))
axs[0].plot(t_x, x)
axs[0].set_title('Sinal de entrada x(t)')
axs[0].set_xlabel('t')
axs[0].set_ylabel('x(t)')
axs[0].grid()

axs[1].plot(t_x, h)
axs[1].set_title('Resposta ao impulso h(t)')
axs[1].set_xlabel('t')
axs[1].set_ylabel('h(t)')
axs[1].grid()

axs[2].plot(t_y, y)
axs[2].set_title('Sinal de saída y(t) obtido por convolução')
axs[2].set_xlabel('t')
axs[2].set_ylabel('y(t)')
axs[2].grid()

plt.tight_layout()
plt.show()
```

```python
t_ = np.linspace(0, 12, 1000)

y = np.where(
    (t >= 0) & (t <= 2),
    np.exp(t) - 1,
    np.where(
        t > 2,
        -np.exp(-2) * np.exp(t) + np.exp(t),
        0
    )
)

plt.figure(figsize=(12, 3))
plt.plot(t_, y)
plt.xlabel('t')
plt.xlim(0,12)
plt.ylabel('y(t)')
plt.title('Resultado da convolução gráfica')
plt.grid(True)

plt.show()
```
